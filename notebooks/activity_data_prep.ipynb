# Activity Data For Simulation:

```python
import pandas as pd
from datetime import datetime as dt

# read the 'activity_data.csv' file from 'data/processed/' folder, and parse the 'date' column as datetime with format 'YYYY-MM-DD'
grouped_df = pd.read_csv('../data/processed/activity_data.csv', parse_dates=['date'], date_parser=lambda x: dt.strptime(x, '%Y-%m-%d'))
grouped_df
```

```python
# drop the shipments column:
grouped_df.drop(columns=['shipments'], inplace=True)
grouped_df
```

```python
grouped_df.info()
```

```python
# change the data type of the date column to datetime:
grouped_df['date'] = pd.to_datetime(grouped_df['date'])

# change the data type of the 'qty', 'inv_qty', and 'items_in_shipments' columns to int:
grouped_df['qty'] = grouped_df['qty'].astype(float)
grouped_df['inv_qty'] = grouped_df['inv_qty'].astype(float)
grouped_df['items_in_shipments'] = grouped_df['items_in_shipments'].astype(float)
grouped_df.info()
```

## Inbound shipments DF:

Start by transfering returned items (negative values) from the `qty` column to the `items_in_shipment` column

```python
# show rows with negative values in the 'qty' or 'inv_qty' columns:
grouped_df[grouped_df['qty'] < 0]
```

```python
# show rows with negative values in the 'qty' or 'inv_qty' columns with and 'inv_qty' bigger than 1:
grouped_df[(grouped_df['qty'] < 0) & (grouped_df['inv_qty'] > 1)]
```

As we can see, we have both cases which are easier to handle by determining that if we have only 1 invoice, then it's obviously only returned items, and days in which we have more then one invoice, so we can't know for sure whether or not we have both returned items and new shipments.

Therefore, we will handle both cases as returns only, by passing the values from the `qty` to the `items_in_shipment` column, and by setting the `inv_qty` column to 0.

```python
# for every row with negative values in the 'qty' column, add the absolute value of the 'qty' to the 'items_in_shipment' column and set the 'qty' and 'inv_qty' values to 0:
grouped_df.loc[grouped_df['qty'] < 0, 'items_in_shipments'] = grouped_df['items_in_shipments'] + abs(grouped_df['qty'])
grouped_df.loc[grouped_df['qty'] <= 0, 'inv_qty'] = 0
grouped_df.loc[grouped_df['qty'] < 0, 'qty'] = 0
```

```python
# Check all values were fixed:
# show rows with negative values in the 'qty' or 'inv_qty' columns:
grouped_df[grouped_df['qty'] < 0]
```

```python
# create a new shipments_df DataFrame from the grouped_df DataFrame, include the 'uuid', 'date', and 'items_in_shipments' columns:
shipments_df = grouped_df[['uuid', 'date', 'items_in_shipments']]
# filter rows where the 'items_in_shipments' column is zero:
shipments_df = shipments_df[shipments_df['items_in_shipments'] != 0]
shipments_df
```

```python
# show all shipments with 'items_in_shipments' which are not full numbers:
shipments_df[shipments_df['items_in_shipments'] % 1 != 0]
```

```python
# round up the 'items_in_shipments' column values to the nearest integer using np.ceil() function:
import numpy as np
shipments_df['items_in_shipments'] = np.ceil(shipments_df['items_in_shipments'])
shipments_df[shipments_df['items_in_shipments'] % 1 != 0]
```

```python
# rename the 'items_in_shipments' column to 'quantity':
shipments_df.rename(columns={'items_in_shipments': 'quantity'}, inplace=True)
# change the data type of the 'quantity' column to int:
shipments_df['quantity'] = shipments_df['quantity'].astype(int)
# change the column order to 'date', 'uuid', 'quantity':
shipments_df = shipments_df[['date', 'uuid', 'quantity']]
# sort by 'date', 'uuid' columns:
shipments_df.sort_values(by=['date', 'uuid'], inplace=True)
# reset the index:
shipments_df.reset_index(drop=True, inplace=True)
shipments_df
```

```python
# check for duplicates:
shipments_df[shipments_df.duplicated()]
```

```python
# check if there are shipments with the same 'date' and 'uuid' but different 'quantity':
shipments_df[shipments_df.duplicated(subset=['date', 'uuid'], keep=False)]
```

```python
# export the shipments_df DataFrame to '../src/data/shipments.csv' file:
shipments_df.to_csv('../src/data/shipments.csv', index=False)
```

## Create the dates list:

```python
# create a new DataFrame called 'dates_df' by grouping the 'date' column of 'grouped_df' by unique values, and summing the 'Inv Qty' column:
dates_df = grouped_df[['date', 'inv_qty']].groupby(['date']).sum().reset_index()
dates_df
```

```python
# add a new column called 'short_day' which contains True if the 'date' values represents a Friday or Saturday, and False otherwise:
dates_df['short_day'] = dates_df['date'].apply(lambda date: date.weekday() in [4, 5])
```

```python
# export the dates_df DataFrame's 'date' and 'short_day' columns to '../src/data/dates.csv' file:
dates_df[['date', 'short_day']].to_csv('../src/data/dates.csv', index=False)
```

## Create the orders DF:

```python
minutes_in_hour = 60
short_day_work_time = (11 - 8) * minutes_in_hour
long_day_work_time = ((16 - 13) + (12 - 8)) * minutes_in_hour
# calculate the total work time for each day:
dates_df['work_time'] = dates_df['short_day'].apply(lambda short_day: short_day_work_time if short_day else long_day_work_time)
# calculate the average time per invoice:
dates_df['avg_time_per_inv'] = dates_df['work_time'] / dates_df['inv_qty']
# calculate the poission lambda:
dates_df['lambda'] = dates_df['inv_qty'] / dates_df['work_time']
# calculate an improved lambda by multiplying the original lambda by 1.5:
dates_df['improved_lambda'] = dates_df['lambda'] * 1.5
dates_df
```

Fix problematic values:

```python
# show me all rows in the grouped_df DataFrame where the 'inv_qty' is bigger than the 'qty':
grouped_df[grouped_df['inv_qty'] > grouped_df['qty']]
```

```python
# show all rows in which 'qty' isn't a full number:
grouped_df[grouped_df['qty'] % 1 != 0]
```

```python
# round up all values in the 'qty' column:
grouped_df['qty'] = grouped_df['qty'].apply(np.ceil)
# if the 'qty' is not 0, but smaller then 'inv_qty', then make 'qty' equal to 'inv_qty':
grouped_df.loc[(grouped_df['qty'] != 0) & (grouped_df['qty'] < grouped_df['inv_qty']), 'qty'] = grouped_df['inv_qty']
```

```python
# add rows to the orders_df DataFrame by iterating over the rows of the grouped_df DataFrame:
import numpy as np

orders = []

for index, row in grouped_df.iterrows():
    # calculate the number of orders to generate for the current row:
    orders_to_generate = row['inv_qty']
    items_left_to_ship = row['qty']
    # iterate over the number of orders to generate:
    # create a list of orders_to_generate length, with random ints which sum to items_left_to_ship:
    while orders_to_generate > 0:
        if orders_to_generate == items_left_to_ship:
            number_of_items = 1
        elif orders_to_generate == 1:
            number_of_items = items_left_to_ship
        else:    
            # create a number_of_items which is a random number between 1 and the number of items left to ship:
            number_of_items = np.random.randint(1, items_left_to_ship - orders_to_generate + 1)
        
        # update parameters:
        items_left_to_ship -= number_of_items
        orders_to_generate -= 1
        
        # create a new order with the current row's 'uuid', 'date', and 'number_of_items':
        order = {
            'uuid': row['uuid'],
            'date': row['date'],
            'qty': number_of_items
        }
        orders.append(order)

# create the orders_df based on the orders list:
orders_df = pd.DataFrame(orders)
orders_df
```

```python
# merge the orders_df with the dates_df on the 'date' column:
orders_dates_df = orders_df.merge(dates_df[['date', 'avg_time_per_inv', 'short_day', 'work_time']], on='date')
orders_dates_df['beta'] = orders_dates_df['avg_time_per_inv'] / 1.25
orders_dates_df
```

We'll now add timestamps for each order which came in, so it can be simulated later on.

Before adding the timestamps, we'll randomize the order in which the orders came within each date, so we don't have streaks of the same UUID.

```python
orders_dates_df = orders_dates_df.groupby('date').apply(lambda x: x.sample(frac=1)).reset_index(drop=True)
# group by 'date', and add a new column called 'time_between_orders' which uses an exponential distribution with the 'beta' value to create an array the size on of the length of the group:
orders_dates_df['time_between_orders'] = orders_dates_df.groupby('date')['beta'].transform(lambda x: np.random.exponential(x.iloc[0], len(x)))
# add a new column called 'time_between_orders' which is the cumulative sum of the 'time_between_orders' column:
orders_dates_df['minutes_from_start_of_day'] = orders_dates_df.groupby('date')['time_between_orders'].transform(lambda x: x.cumsum())
orders_dates_df
```

```python
# 4. add a new column called 'timestamp' which contains the 'date' column from the orders_df, convert it to a datetime object, add 8 hours to it, and then add the 'minutes_from_start_of_day' column to it:
orders_dates_df['timestamp'] = orders_dates_df['date'].apply(lambda date: pd.to_datetime(date) + pd.Timedelta(hours=8)) + pd.to_timedelta(orders_dates_df['minutes_from_start_of_day'], unit='m')
orders_dates_df
```

Fix problematic timestamps:

```python
# round the 'timestamp' value to the nearest second:
orders_dates_df['timestamp'] = orders_dates_df['timestamp'].dt.round('1s')
orders_dates_df
```

```python
# if the row contains 'short_day' == True, then take every datetime object in the 'timestamp', check if it is after 11:00, and if it is, make it 11:00:
orders_dates_df.loc[orders_dates_df['short_day'] == True, 'timestamp'] = orders_dates_df[orders_dates_df['short_day'] == True]['timestamp'].apply(lambda timestamp: timestamp.replace(hour=11, minute=0) if timestamp.hour > 11 else timestamp)

# 6. if the row contains 'short_day' == False, then check if the datetime object in the 'timestamp' column is after 12:00, if yes, add 1 hour to it. If it passes 16:00, then make it 16:00:
# orders_dates_df.loc[orders_dates_df['short_day'] == False, 'timestamp'] = orders_dates_df[orders_dates_df['short_day'] == False]['timestamp'].apply(lambda timestamp: timestamp + pd.Timedelta(hours=1) if timestamp.hour > 12 else timestamp)
# orders_dates_df.loc[orders_dates_df['short_day'] == False, 'timestamp'] = orders_dates_df[orders_dates_df['short_day'] == False]['timestamp'].apply(lambda timestamp: timestamp.replace(hour=16, minute=0) if timestamp.hour > 16 else timestamp)
# orders_dates_df
```

```python
# show all rows with 'short_day' == False and 'timestamp' > 16:00:
orders_dates_df[(orders_dates_df['short_day'] == False) & (orders_dates_df['timestamp'].dt.hour > 16)]
```

```python
# add one hour to all rows with 'short_day' == False and 'timestamp' after 12:00:
orders_dates_df.loc[(orders_dates_df['short_day'] == False) & (orders_dates_df['timestamp'].dt.hour >= 12), 'timestamp'] = orders_dates_df[(orders_dates_df['short_day'] == False) & (orders_dates_df['timestamp'].dt.hour >= 12)]['timestamp'].apply(lambda timestamp: timestamp + pd.Timedelta(hours=1))
# show all rows with 'short_day' == False and 'timestamp' between 12:00 and 13:00:
orders_dates_df[(orders_dates_df['short_day'] == False) & (orders_dates_df['timestamp'].dt.hour >= 12) & (orders_dates_df['timestamp'].dt.hour < 13)]
```

```python
# prepare for export:
final_orders_df = orders_dates_df.copy()
# sort the rows by 'timestamp':
final_orders_df.sort_values(by='timestamp', inplace=True)

# reset the index:
final_orders_df.reset_index(drop=True, inplace=True)

# add a new column called 'order_id' which contains a unique id for each order:
final_orders_df['order_id'] = final_orders_df.index + 1

# Drop unnecessary columns
final_orders_df.drop(['date', 'work_time', 'avg_time_per_inv', 'short_day', 'beta', 'time_between_orders', 'minutes_from_start_of_day'], axis=1, inplace=True)
# Reorder the columns
final_orders_df = final_orders_df[['order_id', 'timestamp', 'uuid', 'qty']]
# rename the 'qty' column to 'quantity':
final_orders_df.rename(columns={'qty': 'quantity'}, inplace=True)
final_orders_df
```
